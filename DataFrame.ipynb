```python
import pandas as pd
```

```python
# List of names
lst1 = ['Mariam', 'Hoda', 'Alaa', 'Gehad','Ahmad', 'Mohamad', 'Omar']
df = pd.DataFrame(lst1)
print(df)
```

```python
# List of their ages
lst2 = [21, 18, 22, 21, 31, 23, 19]

# We use zip() function to create an iterator 
# that will aggregate elements from two or more iterables.

# Calling DataFrame constructor after zipping
# And specify both lists, with columns 
df = pd.DataFrame(zip(lst1, lst2), columns =['Name', 'Age'])
df
```

```python
# We can use one multidimensional list instead of using number of lists
lst3 = [['Mariam', 'Ahmad', 21], ['Hoda', 'Ali', 18],
       ['Alaa', 'Adel', 22], ['Gehad', 'Mosa', 21.5], ['Ahmad', 'Khabeer', 31],
       ['Mohamad', 'Ahmad', 23], ['Omar', 'Al-Saidi', 19.5]]
# We can change the default rows' numbering by using index=[] in DataFrame
df = pd.DataFrame(lst3, index =['a', 'b', 'c', 'd', 'e', 'f', 'g'], 
columns =['First Name', 'Last Name', 'Age'])
df
```

```python
name = ["Mariam", "Alaa", "Gehad", "Aya"]
age = [21, 22, 21.5, 22]
score = [90, 85, 65, 50]
  
# Dictionary of lists 
dict = {'Name': name, 'Age': age, 'Score': score} 
df = pd.DataFrame(dict)
    
df 
```

```python
#adding a column in the DataFrame
marital_status = ['single', 'single', 'engaged', 'married']    
df['Marital Status'] = marital_status
df
```

```python
# Using insert to add a column to the DataFrame
df.insert(3, "Grade", ['A+', 'A', 'D+', 'D'], True)
df
```

```python
# Deleting or droping a column from the DataFrame
df.drop('Score', inplace = True, axis=1)
df
```

```python
# Renaming some columns in the DataFrame
df.rename(columns = {'Name':'Student_Name', 'Marital Status':'Status'}, inplace = True)
df
```
